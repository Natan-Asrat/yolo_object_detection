```python
from ultralytics import YOLO
import cv2
import cvzone
import math
```

```python
cap = cv2.VideoCapture(0)
cap.set(3, 1280)
cap.set(4, 720)
model = YOLO('./yolo_weights/yolov8n.pt')
cls_names = model.names
```

```python
while True:
    success, img = cap.read()
    results = model(img, stream=True)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            w = x2-x1
            h = y2 - y1
            cvzone.cornerRect(img, (x1, y1, x2, y2))
            confidence = math.ceil(box.conf[0]*100)/100
            cvzone.putTextRect(img, f"{cls_names[int(box.cls[0])]} {confidence}", (max(0, x1),max(35, y1)))
    cv2.imshow('image', img)
    cv2.waitKey(1)
```
